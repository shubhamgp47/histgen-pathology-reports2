```python
import csv
import json

# Input and output file paths
input_csv = "gen_vs_gt.csv"
gt_json = "ground-truths.json"
gen_json = "pred.json"

ground_truths = []
generated_reports = []

# Read the CSV file
with open(input_csv, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        case_id = row["Case ID"].strip() + ".tiff"
        generated_text = row["Generated Reports"].strip()
        ground_truth_text = row["Ground Truths"].strip()
        
        # Append data to respective lists
        ground_truths.append({
            "id": case_id,
            "report": ground_truth_text
        })
        generated_reports.append({
            "id": case_id,
            "report": generated_text
        })

# Write the ground truths JSON
with open(gt_json, "w", encoding="utf-8") as f:
    json.dump(ground_truths, f, indent=2, ensure_ascii=False)

# Write the generated reports JSON
with open(gen_json, "w", encoding="utf-8") as f:
    json.dump(generated_reports, f, indent=2, ensure_ascii=False)

print("Conversion complete. Files saved as:")
print(f"- {gt_json}")
print(f"- {gen_json}")
```
